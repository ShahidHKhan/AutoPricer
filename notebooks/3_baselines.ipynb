```python
from auto_pricer.car import Car

train, val, test = Car.from_hub("ShahidHKhan/cars_lite_processed")
print(f"Train: {len(train):,}, Val: {len(val):,}, Test: {len(test):,}")
```

```python
train_mean = sum(c.price for c in train) / len(train)
print(f"Training mean price: ${train_mean:,.2f}")

def constant_pricer(car):
    return train_mean

evaluate(constant_pricer, test)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

CURRENT_YEAR = 2026

def get_car_age(car):
    return CURRENT_YEAR - car.year

def get_text_length(car):
    return len(car.summary) if car.summary else 0

def build_features_v1(cars):
    X = np.array([
        [c.mileage, get_car_age(c), get_text_length(c)]
        for c in cars
    ])
    y = np.array([c.price for c in cars])
    return X, y

X_train, y_train = build_features_v1(train)
X_test, y_test = build_features_v1(test)

lr1 = LinearRegression()
lr1.fit(X_train, y_train)

print("Coefficients (mileage, car_age, text_length):", lr1.coef_)
print("Intercept:", lr1.intercept_)
print("Train R²:", lr1.score(X_train, y_train))
print("Test R²:", lr1.score(X_test, y_test))
```

```python
def linear_regression_pricer_v1(car):
    X = np.array([[car.mileage, get_car_age(car), get_text_length(car)]])
    return lr1.predict(X)[0]

evaluate(linear_regression_pricer_v1, test)
```

```python
def get_horsepower(car):
    return car.horsepower if car.horsepower is not None else 0

def get_has_accidents(car):
    return 1 if car.has_accidents else 0

def build_features_v2(cars):
    X = np.array([
        [c.mileage, get_car_age(c), get_text_length(c), get_horsepower(c), get_has_accidents(c)]
        for c in cars
    ])
    y = np.array([c.price for c in cars])
    return X, y

X_train2, y_train2 = build_features_v2(train)
X_test2, y_test2 = build_features_v2(test)

lr2 = LinearRegression()
lr2.fit(X_train2, y_train2)

print("Coefficients (mileage, car_age, text_length, horsepower, has_accidents):", lr2.coef_)
print("Train R²:", lr2.score(X_train2, y_train2))
print("Test R²:", lr2.score(X_test2, y_test2))

def linear_regression_pricer_v2(car):
    X = np.array([[car.mileage, get_car_age(car), get_text_length(car), get_horsepower(car), get_has_accidents(car)]])
    return lr2.predict(X)[0]

evaluate(linear_regression_pricer_v2, test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

train_docs = [c.summary for c in train]
test_docs = [c.summary for c in test]

vectorizer = CountVectorizer(max_features=5000, stop_words="english")
X_train_text = vectorizer.fit_transform(train_docs)
X_test_text = vectorizer.transform(test_docs)

y_train_text = np.array([c.price for c in train])
y_test_text = np.array([c.price for c in test])

lr_text = LinearRegression()
lr_text.fit(X_train_text, y_train_text)

print("Train R²:", lr_text.score(X_train_text, y_train_text))
print("Test R²:", lr_text.score(X_test_text, y_test_text))

def text_linear_regression_pricer(car):
    X = vectorizer.transform([car.summary])
    return lr_text.predict(X)[0]

evaluate(text_linear_regression_pricer, test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_text = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_text.fit(X_train_text, y_train_text)

print("Train R²:", rf_text.score(X_train_text, y_train_text))
print("Test R²:", rf_text.score(X_test_text, y_test_text))

def random_forest_text_pricer(car):
    X = vectorizer.transform([car.summary])
    return rf_text.predict(X)[0]

evaluate(random_forest_text_pricer, test)
```

```python
from xgboost import XGBRegressor

xgb_text = XGBRegressor(n_estimators=200, max_depth=6, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_text.fit(X_train_text, y_train_text)

print("Train R²:", xgb_text.score(X_train_text, y_train_text))
print("Test R²:", xgb_text.score(X_test_text, y_test_text))

def xgb_text_pricer(car):
    X = vectorizer.transform([car.summary])
    return xgb_text.predict(X)[0]

evaluate(xgb_text_pricer, test)
```

```python
from scipy.sparse import hstack

def build_numeric_only(cars):
    return np.array([
        [c.mileage, get_car_age(c), get_horsepower(c), get_has_accidents(c)]
        for c in cars
    ])

X_train_num = build_numeric_only(train)
X_test_num = build_numeric_only(test)

X_train_combined = hstack([X_train_text, X_train_num])
X_test_combined = hstack([X_test_text, X_test_num])

xgb_combined = XGBRegressor(n_estimators=200, max_depth=6, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_combined.fit(X_train_combined, y_train_text)

print("Train R²:", xgb_combined.score(X_train_combined, y_train_text))
print("Test R²:", xgb_combined.score(X_test_combined, y_test_text))

def xgb_combined_pricer(car):
    X_text = vectorizer.transform([car.summary])
    X_num = np.array([[car.mileage, get_car_age(car), get_horsepower(car), get_has_accidents(car)]])
    X = hstack([X_text, X_num])
    return xgb_combined.predict(X)[0]

evaluate(xgb_combined_pricer, test)
```
